```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    "../data/wheat_price_monthly.csv",
    parse_dates=["date"]
)

df = df.sort_values("date").set_index("date")
```

```python
plt.figure(figsize=(10,4))
plt.plot(df.index, df["price_usd_ton"])
plt.title("Global Wheat Price (Monthly)")
plt.tight_layout()
plt.savefig("../outputs/figures/wheat_price_timeseries.png")
plt.close()
```
